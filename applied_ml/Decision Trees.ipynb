```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
# Cancer dataset
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
(X_cancer, y_cancer) = load_breast_cancer(return_X_y = True) 
```

```python
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,
random_state = 3)

dtc = DecisionTreeClassifier().fit(X_train, y_train)

print(f'Accuracy of Decision Tree classifier on train data: {dtc.score(X_train, y_train)}')
print(f'Accuracy of Decision Tree classifier on test data: {dtc.score(X_test, y_test)}')
```

Setting max decision tree depth to help avoid overfitting

```python
dtc2 = DecisionTreeClassifier(max_depth = 3).fit(X_train, y_train)

print(f'Accuracy of Decision Tree classifier on train data: {dtc2.score(X_train, y_train)}')
print(f'Accuracy of Decision Tree classifier on test data: {dtc2.score(X_test, y_test)}')
```
