```python
%matplotlib inline
%matplotlib ipympl
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import pandas as pd
```

```python
# To access matplotlib's interactivity options we need to go 
# down to the canvas layer
data = np.random.rand(10)

plt.figure()
plt.plot(data)

# here we'll deal with mouse events
def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title(f'Event at pixels {event.x},{event.y} and data {event.xdata},{event.ydata}')

# connect event to an event listener
plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
# Create a pick event

# create list of countries in randomized order
from random import shuffle

origins = ['China', 'Brazil', 'India', 'USA', 'Canada', 'UK',
'Germany', 'Iraq', 'Chile', 'Mexico']
shuffle(origins)

# dataframe with random values for height and weight
df = pd.DataFrame({'height':np.random.rand(10), 'weight':np.random.rand(10), 
'origin': origins})
df
```

```python
plt.figure()

# the picker argument informs matplotlib that the mouse
# doesn't have to click directly on a rendered object, and in 
# this case it can be up to 5 pixels away
plt.scatter(df['height'],df['weight'], picker = 5)
plt.gca().set_xlabel('Height')
plt.gca().set_ylabel('Weight')

plt.show()
```

```python
# now we'll create a function called onpick that takes the event
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title(f'Selected item came from {origin}')

plt.figure()
plt.scatter(df['height'],df['weight'], picker = 5)
plt.gca().set_xlabel('Height')
plt.gca().set_ylabel('Weight')

plt.show()

plt.gcf().canvas.mpl_connect('pick_event', onpick)
```
