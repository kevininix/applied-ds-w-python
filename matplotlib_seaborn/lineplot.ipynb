```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create test data
linear_data = np.array([1,2,3,4,5,6,7,8])
quadratic_data = linear_data**2

# Plot
plt.figure()

# The function assumes the data to be the x-axis and knows
# the two variables are two different data series and colors
# them appropiately 
plt.plot(linear_data, '-o', quadratic_data, '-o')
```

```python
# Matplotlib has its own minilanguage for how to represent
# datapoints
plt.plot([22,44,55], '--r')
```

```python
# The labelling functions work the same
plt.figure()

plt.plot(linear_data, '-o', quadratic_data, '-o')
plt.plot([22,44,55], '--r')

plt.title('A title')
plt.xlabel('Some data')
plt.ylabel('Some other data')  
plt.legend(['Baseline', 'Competition', 'Us'], loc = 2)
```

```python
# Fill between function

plt.figure()

plt.plot(linear_data, '-o', quadratic_data, '-o')

plt.title('A title')
plt.xlabel('Some data')
plt.ylabel('Some other data')  
plt.legend(['Baseline', 'Competition',], loc = 2)

# Let's say we want to highlight the difference between green
# and the blue curves
plt.gca().fill_between(range(len(linear_data)), linear_data,
quadratic_data, facecolor = 'blue', alpha = 0.25) 
```

```python
# Let's change the x-axis to a series of datetime instances

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')
plt.figure(figsize= (10,5))

plt.plot(observation_dates, linear_data, '-o', observation_dates ,quadratic_data, '-o')
#plt.tight_layout()
```

```python
# That didn't work out as we wanted it, so we'll make use of a
# helper function from Pandas to convert numpy dates into standard
# library dates
import pandas as pd

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')

# Use map() function from the standard library
observation_dates = map(pd.to_datetime, observation_dates)

# We'll get an bc map() creates an iterator and matplotlib can't handle that
plt.figure(figsize = (10,5))

plt.plot(observation_dates, linear_data, '-o', observation_dates ,quadratic_data, '-o')
```

```python
# We need to convert it into a list first before plotting 
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))

plt.figure()
plt.plot(observation_dates, linear_data, '-o', observation_dates ,quadratic_data, '-o')
```

```python
# Looks all crammed up w/o the figsize, let's look at altenatives to this
# Let's try changing the x-axis directly

plt.figure()
plt.plot(observation_dates, linear_data, '-o', observation_dates ,quadratic_data, '-o')

x = plt.gca().xaxis 

for item in x.get_ticklabels():
    item.set_rotation(45)
```

```python
# We can also adjust the subplot to make the text fit better
plt.figure()
plt.plot(observation_dates, linear_data, '-o', observation_dates ,quadratic_data, '-o')

x = plt.gca().xaxis 

for item in x.get_ticklabels():
    item.set_rotation(45)

plt.subplots_adjust(bottom = 0.3)
```

```python
# Matplotlib allows for the use of LaTeX
plt.figure()
plt.plot(observation_dates, linear_data, '-o', observation_dates ,quadratic_data, '-o')

x = plt.gca().xaxis 

for item in x.get_ticklabels():
    item.set_rotation(45)

ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_title('Quadratic ($x^2$) vs. Linear ($x$) perfomance')
```
