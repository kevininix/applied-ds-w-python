```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data **2

plt.figure()

plt.subplot(1,2,1)
plt.plot(linear_data, '-o')

plt.subplot(1,2,2)
plt.plot(exponential_data, '-o')
```

```python
# We can use the subplot() function to put the exponential data
# with the linear data if we want
plt.figure()

plt.subplot(1,2,1)
plt.plot(linear_data, '-o')

plt.subplot(1,2,2)
plt.plot(exponential_data, '-o')

plt.subplot(1,2,1)
plt.plot(exponential_data, '-o')
```

```python
# This illustrates a problem of scale, since in the previous graph
# it looked like if the linear data and exponential data had roughly 
# the same area underneath

# To avoid creating misleading plots we must lock the axes between plots
plt.figure()

# We can create a new subplot for the left and right sides and store them
# in variables 
ax1 = plt.subplot(1,2,1)
plt.plot(linear_data, '-o')
# Now we state explicitely that we want them to share the same axis
ax2 = plt.subplot(1,2,2, sharey = ax1)
plt.plot(exponential_data, '-x')
```

```python
# We can also use the subplots() [note the plural], to get many
# axes at once
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True,
sharey = True)

ax5.plot(linear_data, '-o')
```

```python
# Notice that only the bottom and left-most plots have
# their x and y axes on, to turn in on for the rest we
# must iterate over the axes objects
for ax in plt.gcf().get_axes():
    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_visible(True)

fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex=True,
sharey = True)

ax5.plot(linear_data, '-o')
```
