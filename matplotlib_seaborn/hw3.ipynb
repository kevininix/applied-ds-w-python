```python
%matplotlib inline
import matplotlib.pyplot as plt
%matplotlib ipympl
import matplotlib.animation as animation
import numpy as np
import pandas as pd
```

```python
normal = np.random.normal(-2.5, 1,10000)
gamma = np.random.gamma(2, 1.5, 10000)
exponential = np.random.exponential(2, 10000)+7
uniform = np.random.uniform(14,20, 10000)

fig, axs = plt.subplots(2,2)

axs[0,0].hist(normal, density  = True, bins = 20, color ='#EB3866', alpha = 0.8)
axs[0,1].hist(gamma, density  = True, bins = 20, color ='#4B6FEB', alpha = 0.8)
axs[1,0].hist(exponential, density  = True, bins = 20, color ='#E8AB02', alpha = 0.8)
axs[1,1].hist(uniform, density  = True, bins = 20, color ='#3DEB55', alpha = 0.8)

borders = ['top', 'right']

axs[0,0].spines[borders].set_visible(False)
axs[0,1].spines[borders].set_visible(False)
axs[1,0].spines[borders].set_visible(False)
axs[1,1].spines[borders].set_visible(False)

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, 
top=0.9, wspace=0.4, hspace=0.4)

fig.suptitle('Understanding Distributions Through Sampling')
plt.show()

```
