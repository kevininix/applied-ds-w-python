```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Heatmaps are a way of visualizing 3d dimentional data by
# taking advantage of the proximity of those dimensions
y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000)
x = np.random.random(size = 10000)

plt.figure()
_ = plt.hist2d(x,y, bins = 25)
```

```python
# with more bins 
plt.figure()
_ = plt.hist2d(x,y, bins = 100)
```

```python
# adding legends
plt.figure()
_ = plt.hist2d(x,y, bins = 100)
plt.colorbar()
plt.show()
```
