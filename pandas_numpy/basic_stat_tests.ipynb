```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Hypothesis testing example with grade data
df = pd.read_csv('datasets/grades.csv')
df.head()
```

```python
# Summary statistics
print(f'There are {df.shape[0]} rows and {df.shape[1]} columns')
```

```python
# Lets segment the data in two: those who finished the assignment
# by Dec 2015, and those who finished after said  date 
early_finishers = df[pd.to_datetime(df['assignment1_submission']) < '2016']
early_finishers.head()
```

```python
#late_finishers = df[pd.to_datetime(df['assignment1_submission']) > '2015-12'] <- alternative
late_finishers = df[~df.index.isin(early_finishers.index)]
 
late_finishers.head()
```

```python
# Compare the mean between these populations 
print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

```python
# Is this difference significant? Let's do hypothesis testing

# The scipy function ttest_ind() retunr the t-test and p-value
# on two unrelated (independent) populations 
from scipy.stats import ttest_ind
```

```python
# The p-value is way above our significant value of 0.05 which means
# we can't reject the null hypothesis, i.e the difference between grades
# (for the 1st assignment) is not significant
ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

```python
# Let's check if this is the case for the other assignments

# So we're unable to reject the null hypothesis for all groups, meaning
# that we can't say that there are any significant differences between
# early and late finishers
print(ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print(ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```

```python
# P-values alone might not tell the full story however, so making use
# of confidence intervals and bayesian analyses is also a good idea

# One of the issues with p-values is that as you run more tests you
# are likely to get a value that is statistically significant 
```

```python
# Example using a dataframe with 100 columns each with 100 values

df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
```

```python
# Are these two dataframes the same? Or, for a given row in df1
# is it the same as that same row in df2?
# Let's assume a significance level of \alpha = 0.1

# Let's compare each column in df1 to the same numbered column in 
# df2 
def test_columns(alpha = 0.1):

    # number of columns that differ
    num_diff = 0

    for col in df1.columns:

        teststat,pval = ttest_ind(df1[col],df2[col])

        if pval <= alpha:
            print(f'Column {col} is statistically significantly different at alpha = {alpha}, pval = {pval}')
            num_diff += 1
    
    # summary statistics
    print(f'Total number of different columns was {num_diff}, which is {(float(num_diff/len(df1.columns))*100)}%')

test_columns()
```

```python
# Notice that the percentage of different columns is ver close to
# our chosen significance level of 10%, this is because all ttest 
# does is check if two sets are similar given some level of confidence

# The more random comparisons you do, the more will just happen to be
# the same by chance. In this example we used 100 columns, so we would 
# expect there to be roughly 10 of them if \alpha = 0.10
```

```python
# We can test some other alpha values as well 
test_columns(0.05)
```
