```python
import pandas as pd
```

### Querying

```python
# Boolean masking is at the heart of querying in Pandas
# A boolean mask is an n-dimensional array where each of its
# members is either true or false
df = pd.read_csv('datasets/Admission_Predict.csv', index_col = 0)

# Let's clean this dataset
df.columns = [x.lower().strip() for x in df.columns]
df.head()
```

```python
# Boolean masks are created by applying operators directly on 
# the Pandas Series or DataFrame and returning only the True
# values

# Let's filter out students with a chance of admit < 0.7
admit_mask = df['chance of admit'] > 0.7
admit_mask
```

```python
# Now we lay this mask on top the data to hide False values
# using the .where() function
df.where(admit_mask).head()
```

```python
# Now we use the .dropna() function to get rid of missing data
df.where(admit_mask).dropna().head()
```

```python
# You can do the same as the where().dropna() combo using 
# the indexing operator
df[df['chance of admit'] > 0.7].head()
```

```python
# For combining multiple boolean masks you can use the 
# "&" and "|" ('or') logical operators

# Parentheses are important since they indicate order of
# operations
df[(df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)].head()
```

```python
# An alternative is to use built-in functions of comparison
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```python
# Since they're functions you can chain them
df['chance of admit'].gt(0.7).lt(0.9)
```

### Indexing

```python
# The set_index() function is a destructive process, and 
# it doesn't keep the currrent index 
df.head()
```

```python
# Let's change the index column to chance of admision but
# conserve the Serial No. column

# We need to make a copy of Serial No.
df['Serial Number'] = df.index

# No we set the index to another column
df = df.set_index('chance of admit')
df.head()
```

```python
# To get rid of the index column we use the reset_index()
# function, this promotes it to a column and sets in the default
# numbered index list
df = df.reset_index()
df.head()
```

```python
# In Pandas we can also do multi-level indexing, this useful
# in e.g geographical data divided by regions or demograpics

# Let's see this with some demographical data
df = pd.read_csv('datasets/census.csv')
df.head()
```

```python
# This dataset has two levels of summarization:
# country level and estate level

# To see the unique values of a column use the .unique()
# function
df['SUMLEV'].unique()
```

```python
# Let's exclude all the state-level summaries a 
# keep just the county-level ones
df = df[df['SUMLEV'] == 50]
df.head()
```

```python
# Let's focus only in the total populaton and total
# births
columns_to_keep = ['STNAME', 'CTYNAME', 'BIRTHS2010', 'BIRTHS2011', 'BIRTHS2012', 'BIRTHS2013',
'BIRTHS2014', 'BIRTHS2015', 'POPESTIMATE2010', 'POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013',
'POPESTIMATE2014', 'POPESTIMATE2015']

df = df[columns_to_keep]
df.head()
```

```python
# We can set the index to be a combinarion of state and county
df = df.set_index(['STNAME', 'CTYNAME']) 
df.head()
```

```python
# To query this DataFrame we must provide the index
# by order of generality

# To watch the population result for Washtenaw county in
# Michigan, the 1st arg is Michigan and the 2nd one is
# Washtenaw county
df.loc['Michigan', 'Washtenaw County']
```

```python
# To compare two counties we pass a list of tuples, each
# Containing two elements for state and county
df.loc[[('Michigan', 'Washtenaw County'), ('Michigan', 'Wayne County')]]
```

```python
# If we tranpose this DataFrame we'll have hierarchilal
# columns labels
df.loc[[('Michigan', 'Washtenaw County'), ('Michigan', 'Wayne County')]].T.head()
```
