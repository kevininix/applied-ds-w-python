```python
import pandas as pd
```

```python
# Let's clean the dataset of President names scrapped from
# Wikipedia
df = pd.read_csv('datasets/presidents.csv')
df.head()
```

```python
# Separate names into first and last names and get rid of 
# the footnotes in the Born column
df['First'] = df['President']
df['First'] = df['First'].replace(to_replace = '[ ].*', value = "", regex = True)
df.head()
```

```python
# This works but takes time and the code it's kinda ugly
# Let's drop it and try better methods
del df['First']
```

```python
# The more general alternative is to use the apply() function

# Let's write a function that splits a string into two pieces
def split_name(row):
    row['First'] = row['President'].split(" ")[0]
    row['Last'] = row['President'].split(" ")[-1]
    return row

df= df.apply(split_name, axis = 'columns')
df.head()
```

```python
# Another alternative is to use the .str.extract() method

# Let's drop First and Last to see how to use it
del df['First']
del df['Last']
```

```python
# Extract takes a regex and requires you to specify capture 
# groups corresponding to the output columns
pattern = '(^[\w]*)(?:.* )([\w]*$)'

df['President'].str.extract(pattern).head()
```

```python
# If we name the groups the names will be read as column names
pattern = '(?P<First>^[\w]*)(?:.* )(?P<Last>[\w]*$)'

names = df['President'].str.extract(pattern)
names.head()
```

```python
# Now just add them to the DataFrame
df['First'] = names['First']
df['Last'] = names['Last']

df.head()
```

```python
# Let's clean the Born column now
df['Born'] = df['Born'].str.extract('([\w]{3} [\w]{1,2}, [\w]{4})')
df['Born'].head()
```

```python
# Even tho the data type is object, we could change it
# to something better suited for dates
df['Born'] = pd.to_datetime(df['Born'])
df['Born'].head()
```

```python
# This makes it easier to do operations involving dates
# e.g getting every President born during a give time-span
```

```python
df = df.set_index('#')
df.head()
```

```python
df['Age atstart of presidency'] = df['Age atstart of presidency'].str.extract('(.*)(?:[A-Z]{1})')
df['Age atend of presidency'] = df['Age atend of presidency'].str.extract('(.*)(?:[A-Z]{1})')

df = df.rename(columns = {'Age atstart of presidency': 'Age at start of presidency',
'Age atend of presidency': 'Age at end of presidency', 'Post-presidencytimespan': 'Post-presidency timespan'})

df.head()
```
