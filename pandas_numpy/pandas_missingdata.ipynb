```python
import pandas as pd
```

```python
# Missing data is usually formatted as NaN, NULL, None or N/A 
# although in certain cases it can be formatted as e.g 99

# The read_csv() function lets us specify the form of the missing
# values with the na_values parameter 
df = pd.read_csv('datasets/class_grades.csv')
df.head(10)
```

```python
# We can use the .isnull() function to create a boolean mask
# This can be useful for processing rows based on certain
# columns of data

mask = df.isnull()
mask.head(10)
```

```python
# To drop rows with missing data we can use the .dropna()
# function
df.dropna().head(10)
```

```python
# We can also fill the missing data with other values
# using the fillna() function
df.fillna(0, inplace = True)
df.head(10)
```

```python
# If white spaces are object of interest, we can turn-off
# white space filtering with the na_filter parameter in
# read_csv(). In data with no NAs, na_filter = False can 
# improve perfomance of readinf large files
```

```python
# Missing values sometimes carry information

# For this dataset NaN means that no changes have been made
# since the playback has been paused
df = pd.read_csv('datasets/log.csv')
df.head(10)
```

```python
# We can also use the parameter() method to do foward and 
# backwrad filling. Data has to be sorted before using it.
# ffill updates NA values with the value of the previous valid row
# bfill updates NA values with the value of the next valid row
```

```python
# In Pandas we can sort by index or by values
df = df.set_index('time')
df = df.sort_index()
df.head()
```

```python
# Notice, however, that the index isn't unique
# Let's reset the index and use multi-level indexing
# for time AND user 
df = df.reset_index()
df = df.set_index(['time', 'user'])
df.head()
```

```python
# With the data now correctly indexed, we can use ffill
df = df.fillna(method = 'ffill')
df.head()
```

```python
# The replace() function allows for several types of
# customized fill-ins: value-to-value, list, dictionary, regex.
df = pd.DataFrame({'A':[1,1,2,3,4], 'B':[3,6,3,8,9], 'C':['a','b','c','d','e']})
df
```

```python
# Let's replace 1's with 100
df.replace(1, 100)
```

```python
# To replace several values
df.replace([1, 3], [100, 300])
```

```python
# Pandas replacement also supports regex
df = pd.read_csv('datasets/log.csv')
df.head()
```
