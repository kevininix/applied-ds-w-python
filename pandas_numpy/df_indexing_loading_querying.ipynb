```python
import pandas as pd
```

### Indexing and loading

```python
# Let's load a .csv file
df = pd.read_csv('datasets/Admission_Predict.csv')
df.head()
```

```python
# Set the Serial number column as the index with index_col
df = pd.read_csv('datasets/Admission_Predict.csv', index_col = 0)
df.head()
```

```python
# Change column names with the rename() function
new_df = df.rename(columns = {'GRE score':'GRE score', 'TOEFL Score':'TOEFL Score',
'University Rating':'University Rating', 'SOP': 'Statement of Purpose',
'LOR':'Letter of Recommendation', 'CGPA':'CGPA', 'Research':'Research',
'Chance of Admit':'Chance of Admit'})

new_df.head()
```

```python
# Why wasn't LOR renamed? 
# First let's make sure we got all the column names correct
new_df.columns
```

```python
# Notice that that there is a space after LOR and Chance of Admit
# This is why LOR didn't change, we wrote the key incorrectly
new_df = new_df.rename(columns = {'LOR ': 'Letter of Recommendation'})
new_df.head()
```

```python
# A more straightfoward method is to use the strip() function, passed
# as the mapper paremeter, to get rid of spaces. We jus need to indicate
# the axis (columns or row/index label)
new_df = new_df.rename(mapper = str.strip, axis = 'columns')
new_df.columns
```

```python
# We can also use the .columns attribute to modify the og dataframe
# by assigning it to a list
# This method is not affected ny subtle errors like extra spaces 
cols = list(df.columns)

# Now we do a list comprehension to turn all names to lowercase
cols = [x.lower().strip() for x in cols]
df.columns = cols

df.head()
```
